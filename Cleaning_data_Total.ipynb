```python
import pandas as pd 
    
# making dataframe 
energy_sources = pd.read_csv("Energy data 1990 - 2020.csv") 
   
# output the dataframe
energy_sources
```

```python
energy_sources.columns
```

```python
#change column names: 

energy_sources.columns = ['Country', 
              'Year', 
              'Region',
       'CO2_emissions',
       'Average_CO2_factor',
       'CO2_PPP',
       'Total_energy_production', 
       'Total_energy_consumption',
       'Share_renewables_electricity_production',
       'Share_electricity_final_energy_consumption',
       'Oil_products_consumption',
       'Refined_oil_products_production',
       'Natural_gas_production',
       'Natural_gas_consumption',
       'Energy_intensity_GDP_PPP',
       'Electricity_production',
       'Electricity_consumption',
       'Coal_lignite_consumption',
       'Share_wind_solar_electricity_production',
       'Crude_oil_production',
       'Coal_lignite_production']
```

```python
#re-order cols
energy_sources = energy_sources[[
    #summary (most important)
    'Country', 
    'Year', 
    'Region',
    'Total_energy_production', 
    'Total_energy_consumption',
    'CO2_emissions',

    #non-renewables
    'Oil_products_consumption',
    'Refined_oil_products_production',
    'Natural_gas_production',
    'Natural_gas_consumption',
    'Coal_lignite_production', 
    'Coal_lignite_consumption',
    'Crude_oil_production',
    
    #renewables
    'Share_wind_solar_electricity_production',
    'Share_renewables_electricity_production',
    'Share_electricity_final_energy_consumption',

    #dont know what this means
    'Average_CO2_factor',
    'CO2_PPP',
    'Energy_intensity_GDP_PPP',
    
    #output of energy
    'Electricity_production',
    'Electricity_consumption']]

energy_sources
```

```python
# remove all n.a. values
energy_sources = energy_sources[energy_sources["Region"] != "n.a."]
energy_sources = energy_sources[energy_sources["Total_energy_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["CO2_emissions"] != "n.a."]
energy_sources = energy_sources[energy_sources["Average_CO2_factor"] != "n.a."]
energy_sources = energy_sources[energy_sources["CO2_PPP"] != "n.a."]
energy_sources = energy_sources[energy_sources["Total_energy_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["Total_energy_consumption"] != "n.a."]
energy_sources = energy_sources[energy_sources["Share_renewables_electricity_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["Share_electricity_final_energy_consumption"] != "n.a."]
energy_sources = energy_sources[energy_sources["Oil_products_consumption"] != "n.a."]
energy_sources = energy_sources[energy_sources["Refined_oil_products_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["Natural_gas_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["Natural_gas_consumption"] != "n.a."]
energy_sources = energy_sources[energy_sources["Energy_intensity_GDP_PPP"] != "n.a."]
energy_sources = energy_sources[energy_sources["Electricity_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["Electricity_consumption"] != "n.a."]
energy_sources = energy_sources[energy_sources["Coal_lignite_consumption"] != "n.a."]
energy_sources = energy_sources[energy_sources["Share_wind_solar_electricity_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["Crude_oil_production"] != "n.a."]
energy_sources = energy_sources[energy_sources["Coal_lignite_production"] != "n.a."]
energy_sources
```

```python
energy_sources.drop(columns=["Country", "Region"])
```

```python
# Convert to datetime format
energy_sources['Year'] = pd.to_datetime(energy_sources['Year'])

# Calculate sum of columns by year
Energy_by_year = energy_sources.groupby(energy_sources['Year'].dt.year).sum()
Energy_by_year
```

```python
Energy_by_year.to_csv("Energy_by_year.csv")
```
